**1. Створення масивів**

Створіть (кожен пункт оцінюється в 0.5 бала):
- Масив зі списку `[5, 15, 25, 35, 45]`
- Масив з нулів розміром 7
- Масив з одиниць розміром 6
- Масив з числами від 10 до 19 використовуючи `np.arange()`

```python
import numpy as np
arr = np.array([5, 15, 25, 35, 45])
arr_zeroes = np.zeros(7)
arr_ones = np.ones(6)
arr_range = np.arange(10, 20, 1)
print("Original array:", arr)
print("Array of zeroes:", arr_zeroes)
print("Array of ones:", arr_ones)
print("Array with range 10 to 19:", arr_range)
```

**2. Базові операції**

Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2
- Додайте до всіх елементів 10
- Піднесіть всі елементи до квадрату
- Обчисліть квадратний корінь з усіх елементів

Виведіть результат виконання кожної операції.

```python
arr2 = np.array([12, 24, 36, 48, 60])
arr_doubled = arr2 * 2
arr_plus_ten = arr2 + 10
arr_squared = np.square(arr2)
arr_sqrt = np.sqrt(arr2)
print("Array after multiply by 2:", arr_doubled)
print("Array after addition of 10:", arr_plus_ten)
print("Array after squaring:", arr_squared)
print("Square root of the array:", arr_sqrt)
```

**3. Статистичні функції**

Створіть Numpy-масив `data` зі значеннями [18, 27, 33, 41, 52, 68, 79].

Обчисліть використовуючи функціонал numpy (кожен пункт оцінюється в 0.5 бала):
- Середнє значення
- Мінімум
- Максимум
- Суму всіх елементів

```python
arr3 = np.array([18, 27, 33, 41, 52, 68, 79])
arr_mean = np.mean(arr3)
arr_min = np.min(arr3)
arr_max = np.max(arr3)
arr_sum = np.sum(arr3)

print("Mean of the array:", arr_mean)
print("Minimum value in the array:", arr_min)
print("Maximum value in the array:", arr_max)
print("Sum of all elements in the array:", arr_sum)
```

**4. Індексація**

Створіть Numpy-масив `arr` зі значеннями [15, 22, 35, 44, 58, 67].

Виведіть (кожен пункт оцінюється в 0.5 бала):
- Перший елемент
- Останній елемент
- Елементи з індексами 1 до 3
- Елементи більші за 40

```python
arr4 = np.array([15, 22, 35, 44, 58, 67])

first_item = arr4[0]
last_item = arr4[-1]
sub_array = arr4[1:4]
elements_greater_than_40 = arr4[arr4 > 40]

print("First item of the array:", first_item)
print("Last item of the array:", last_item)
print("Sub-array (1nd to 3rd items inclusive 3rd):", sub_array)
print("Elements greater than 40:", elements_greater_than_40)
```

**5. Двовимірні масиви**

Задано матрицю 3x3.

Виконайте наступні дії з функціоналом Numpy (кожен пункт оцінюється в 0.5 бала):
- Виведіть форму (shape) матриці
- Виведіть перший рядок
- Виведіть другу колонку
- Обчисліть суму всіх елементів
- Обчисліть середнє всіх елементів

```python
matrix = np.array([[2, 4, 6],
                   [8, 10, 12],
                   [14, 16, 18]])
```

```python
shape = matrix.shape
first_row = matrix[0, :]
second_column = matrix[:, 1]
summ_of_all_elements = np.sum(matrix)
mean_if_all_elements = np.mean(matrix)

print("Shape of the matrix:", shape)
print("First row of the matrix:", first_row)
print("Second column of the matrix:", second_column)
print("Sum of all elements in the matrix:", summ_of_all_elements)
print("Mean of all elements in the matrix:", mean_if_all_elements)
```
