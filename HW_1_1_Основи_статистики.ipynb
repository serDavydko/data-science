**Завдання 3.** Для того ж набору даних в змінній `sum_purchases` обчисліть **медіану** кожним з зазначених методів:  

  3.1. без використання жодних бібліотек, користуючись тільки Python.  
  3.2. використовуючи відповідний метод бібліотеки numpy.  

Виведіть результат обчислень в кожному з випадків на екран. Порівняйте результати зі значенням вибіркового середнього. Напишіть коментар, для яких задач ви б використовували вибіркове середнє, а для яких - медіану?

```python
def calculate_median(data):
    """Median for even and odd lists."""
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        return (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    else:
        return sorted_data[n // 2]

median_python = calculate_median(sum_purchases)
median_numpy = np.median(sum_purchases)

print(f"Median of purchases python: {median_python}")
print(f"Median of purchases numpy: {median_numpy}")
print(f"Results are {'Equal' if median_python == median_numpy else 'Not Equal'}")

print(f"Mean is {mean_python} and Median is {median_python}")
print("In this case mean is higher than median, indicating a right-skewed distribution due to outliers 4310 and 3524.")
print("Use mean when data is normally distributed without outliers, for example heights of people.\nUse median when data has outliers or is skewed, for example house prices in a city.")
```

**Завдання 4**. Обчисліть стандартне відхилення наведеної вибірки в змінній `sum_purchases` кожним з зазначених методів:  

  4.1. без використання жодних бібліотек, користуючись тільки Python.  
  4.2. використовуючи відповідний метод бібліотеки numpy.  

Виведіть результат обчислень в кожному з випадків на екран.

```python
def st_deviation(data, ddof=0):
    """Standard deviation calculation."""
    mean = sum(data) / len(data)
    variance = sum((x - mean) ** 2 for x in data) / (len(data) - ddof)
    return variance ** 0.5

st_deviation_python = st_deviation(sum_purchases, ddof=1)
st_deviation_numpy = np.std(sum_purchases, ddof=1)

print(f"Standard Deviation of sample python: {st_deviation_python}")
print(f"Standard Deviation of sample numpy: {st_deviation_numpy}")
print(f"Results are {'Equal' if st_deviation_python == st_deviation_numpy else 'Not Equal'}")
```

- 10 студентів мають вік 16 років
- 25 студентів - 17 років
- 20 студентів - 18 років
- 2 студенти - 20 років

Обчислість середнє значення для цієї вибірки будь-яким способом - на чистому Python або з numpy - та виведіть результат на екран.

```python
ages_sample = [[16, 10], [17, 25], [18, 20], [20, 2]] #ages and number of peoples

def mean_counter_weighted(data):
    total_weighted_sum = sum(age * count for age, count in data)
    total_count = sum(count for age, count in data)
    return total_weighted_sum / total_count

mean_weighted_python = mean_counter_weighted(ages_sample)
mean_weighted_numpy = np.average(
    [age for age, count in ages_sample],
    weights=[count for age, count in ages_sample]
)
print(f"Mean age weighted by number of people (python): {mean_weighted_python}")
print(f"Mean age weighted by number of people (numpy): {mean_weighted_numpy}")
```

**Завдання 6**. Крім даних про суму покупок, ми зібрали ще дані про кількість відвідувань кожного з кожного покупця, які знаходяться в змінній `visits`. Обчисліть коефіцієнт кореляції Пірсона для суми покупок і кількості відвідувань кожним з зазначених методів:  

  6.1. без використання жодних бібліотек, користуючись тільки Python.  
  6.2. використовуючи відповідний метод бібліотеки numpy.  
  
Виведіть результат на екран. Проаналізуйте результат.

Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]

def pirson_coefficient(x, y):
    """Calculate Pearson correlation coefficient."""
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x_sq = sum(i**2 for i in x)
    sum_y_sq = sum(i**2 for i in y)
    sum_xy = sum(x[i] * y[i] for i in range(n))
    
    numerator = (n * sum_xy) - (sum_x * sum_y)
    denominator = ((n * sum_x_sq - sum_x**2) * (n * sum_y_sq - sum_y**2)) ** 0.5
    
    if denominator == 0:
        return 0
    return numerator / denominator

pirson_coefficient_python = pirson_coefficient(sum_purchases, visits)
pirson_coefficient_numpy = np.corrcoef(sum_purchases, visits)[0, 1]

print(f"Correlation between purchases and visits (python): {pirson_coefficient_numpy}")
print(f"Correlation between purchases and visits (numpy): {pirson_coefficient_python}")

print('Its better to exclude outliers: 4310, 3524 before calculating correlation as they skewed the results significantly.')
```

**Завдання 7**. В попередньому завданні ми помітили, що викиди нам заважають отримати "правдиве" значення кореляції. Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
def remove_outliers(data_x, data_y, threshold_multiplier=2):
    """
    Remove outliers from paired datasets based on threshold in data_x.
    """
    data_x = np.array(data_x)
    data_y = np.array(data_y)
    
    mean_x = np.mean(data_x)
    std_x = np.std(data_x, ddof=1)
    
    threshold = mean_x + threshold_multiplier * std_x
    
    mask = data_x <= threshold
    filtered_x = data_x[mask]
    filtered_y = data_y[mask]
    
    removed_count = len(data_x) - len(filtered_x)
    
    return filtered_x.tolist(), filtered_y.tolist(), threshold, removed_count


filtered_purchases, filtered_visits, threshold, removed = remove_outliers(sum_purchases, visits)

print(f"Threshold for outliers: {threshold:.2f}")
print(f"\nOriginal data size: {len(sum_purchases)}")
print(f"Filtered data size: {len(filtered_purchases)}")
print(f"Removed outliers: {removed}")

pirson_coefficient_no_outliers = np.corrcoef(filtered_purchases, filtered_visits)[0, 1]

print(f"\nCorrelation with outliers: {pirson_coefficient_python:.4f}")
print(f"Correlation without outliers: {pirson_coefficient_no_outliers:.4f}")


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(sum_purchases, visits, alpha=0.6)
plt.title('With Outliers')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits')

plt.subplot(1, 2, 2)
plt.scatter(filtered_purchases, filtered_visits, alpha=0.6, color='green')
plt.title('Without Outliers')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits')

plt.tight_layout()
plt.show()

print("\nAnalysis:")
print(f"After removing outliers, the correlation coefficient changed from {pirson_coefficient_python:.4f} to {pirson_coefficient_no_outliers:.4f}.")
print("The correlation became stronger, indicating a clearer negative linear relationship between the number of visits and the sum of purchases.")
```
